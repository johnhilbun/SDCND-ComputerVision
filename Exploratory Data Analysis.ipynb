## Write a function to display an image and the bounding boxes

Implement the `display_instances` function below. This function takes a batch as an input and display an image with its corresponding bounding boxes. The only requirement is that the classes should be color coded (eg, vehicles in red, pedestrians in blue, cyclist in green).

```python
def display_instances(batch):
    """
    This function takes a batch from the dataset and display the image with 
    the associated bounding boxes.
    """
    # ADD CODE HERE
    # Help : extract image from dataset - https://knowledge.udacity.com/questions/727411
    
    try:
        # color map
        color_map = {1: 'red', 2: 'blue', 4: 'green'}
        
        fig, ax = plt.subplots()
        im = batch["image"].numpy()
        ax.imshow(im)
        
        # draw bounding boxes and add color to different class labels
        for coord, label in zip(batch["groundtruth_boxes"].numpy(),
                               batch["groundtruth_classes"].numpy()):
            xy = (coord[1] * im.shape[1], coord[0] * im.shape[0])
            
            #taken from class notes
            width = (coord[3] - coord[1]) * im.shape[1]
            height = (coord[2] - coord[0]) * im.shape[0]
            
            rec = matplotlib.patches.Rectangle(xy, width, height, linewidth=1,
                                   edgecolor=color_map[label], facecolor='none')
            ax.add_patch(rec)
        plt.axis("off")
    except Exception as err:
        print (err)
 
```

## Display 10 images 

Using the dataset created in the second cell and the function you just coded, display 10 random images with the associated bounding boxes. You can use the methods `take` and `shuffle` on the dataset.

```python
## STUDENT SOLUTION HERE
# Help : extract image from dataset - https://knowledge.udacity.com/questions/727411

plt.figure(figsize=(10,10))

for i in range(10):
    for batch in dataset.shuffle(75, reshuffle_each_iteration=True).take(1):
        display_instances(batch)

plt.show()
```
